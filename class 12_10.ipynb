```python
import numpy as np
x = np.linspace(1, 15, 15)
x
```

```python
y = [5, 12, 17, 20, 28, 25, 47, 42, 50, 57, 61, 63, 81, 89, 93]
import matplotlib.pyplot as plt
plt.scatter(x, y)
```

```python
model1 = np.polyfit(x, y, 2)
print(model1)
yfitted = x**2*model1[0] + x*model1[1] + model1[2]
plt.plot(x, y, "bo")
plt.plot(x, yfitted, "k--")
```

```python
x_matrix = np.array([x**0,x**1,x**2])
```

```python
model2 = np.linalg.lstsq(x_matrix.T, y,rcond=None)
model2
```

Model 1, found by the library code gave the same result as the linear algebra method, model 2, so the code works correctly.
